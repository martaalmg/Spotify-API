```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth


SPOTIPY_CLIENT_ID = '849c1f443d28408a97ba073265cf498c'
SPOTIPY_CLIENT_SECRET = '2f816c2c8e74410eaabe03b4b63d6a95'
SPOTIPY_REDIRECT_URI = 'http://localhost:8888' #If your app name is My Awesome App, a good candidate for the redirect URI could be my-awesome-app-login://callback

# Set up authorization and obtain access token
sp_oauth = SpotifyOAuth(client_id=SPOTIPY_CLIENT_ID,
                        client_secret=SPOTIPY_CLIENT_SECRET,
                        redirect_uri=SPOTIPY_REDIRECT_URI,
                        scope="user-library-read",
                        open_browser=False)

# Get the authorization URL
auth_url = sp_oauth.get_authorize_url()
print(f"Please visit this URL to authorize the application: {auth_url}")

# After user grants permission, capture the authorization code and exchange it for an access token
code = input("Enter the authorization code: ")
token_info = sp_oauth.get_access_token(code)

# Create a Spotify object with the access token
sp = spotipy.Spotify(auth=token_info['access_token'])

# Now you can make requests to the Spotify API with sp
```

```python
playlist_id = '1ZW1qrvSPdiDYT7LImM24l' #My playlist ID

limit = 100
offset = 0

all_tracks_with_features = []

while True:
      playlist = sp.playlist_tracks(playlist_id, limit=limit, offset=offset)
      print(offset)
      if 'items' in playlist and playlist['items']:
            for item in playlist['items']:
                track_id = item['track']['id']
                name = item['track']['name']

                audio_features = sp.audio_features([track_id])[0]
                artists = item['track']['artists']
                artist_genres = []

                if artists:
                  for artist in artists:
                      try:
                        artist_info = sp.artist(artist['id'])
                        artist_genres.extend(artist_info.get('genres', []))
                      except spotipy.SpotifyException as e:
                        print(e)

                if audio_features:
                  track_features = {
                      'id': track_id,
                      'name': name,
                      'acousticness': audio_features['acousticness'],
                      'danceability': audio_features['danceability'],
                      'energy': audio_features['energy'],
                      'instrumentalness': audio_features['instrumentalness'],
                      'key': audio_features['key'],
                      'liveness': audio_features['liveness'],
                      'loudness': audio_features['loudness'],
                      'mode': audio_features['mode'],
                      'speechiness': audio_features['speechiness'],
                      'tempo': audio_features['tempo'],
                      'valence': audio_features['valence'],
                      'genres': artist_genres}

                  all_tracks_with_features.append(track_features)

            # Update offset for the next iteration
            offset += limit

            # Break the loop if all tracks have been retrieved
            if len(playlist['items']) < limit:
                break
```

```python
import pandas as pd
imb = pd.DataFrame(all_tracks_with_features)
imb
```

```python
imb.to_csv('imb.csv', index=False)
```
