```python
grupos_por_categoria = df.groupby('Category')
conjuntos_por_categoria = {nombre_categoria: grupo_df for nombre_categoria, grupo_df in grupos_por_categoria}
```

```python
afro = conjuntos_por_categoria['Afro']
afro
```

```python
afro.describe()
```

```python
Q1 = afro.quantile(0.25)
Q3 = afro.quantile(0.75)
IQR = Q3 - Q1

outliers = ((afro < (Q1 - 1.5 * IQR)) | (afro > (Q3 + 1.5 * IQR))).any(axis=1)
outliers_df = afro[outliers]
outliers_df
```

```python
df_merged = pd.merge(afro, outliers_df, how='left', indicator=True)
afro = df_merged[df_merged['_merge'] == 'left_only']
afro = afro.drop('_merge', axis=1)
afro
```
