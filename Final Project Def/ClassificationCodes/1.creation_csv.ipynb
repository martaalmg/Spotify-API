```python
country_codes = ['AD', 'AR', 'AU', 'AT', 'BE', 'BO', 'BR', 'BG', 'CA', 'CL', 'CO', 'CR', 'CY', 'CZ', 'DK', 'DO', 'EC', 'SV', 'EE', 'FI', 'FR',
'DE', 'GR', 'GT', 'HN', 'HK', 'HU', 'IS', 'ID', 'IE', 'IT', 'JP', 'LV', 'LI', 'LT', 'LU', 'MY', 'MT', 'MX', 'MC', 'NL', 'NZ', 'NI',
'NO', 'PA', 'PY', 'PE', 'PH', 'PL', 'PT', 'SG', 'ES', 'SK', 'SE', 'CH', 'TW', 'TR', 'GB', 'US', 'UY']

categories_set = set()
categories = []
all_playlists = []

#CADA CATEGORIAS EN CADA PA√çS
for country_code in country_codes:
    data = sp.categories(country=country_code, locale=None, limit=20, offset=0)
    country_categories = data['categories']['items']

    for category in country_categories:
        name = category['name']
        category_id = category['id']

        # Verify if the category is not in the set before adding it
        if category_id not in categories_set:
            categories_set.add(category_id)
            categories.append({'name': name, 'id': category_id})
            #print(f"Added: Name: {name}, ID: {category_id}")

# Print the final list of categories
print(categories)
```
